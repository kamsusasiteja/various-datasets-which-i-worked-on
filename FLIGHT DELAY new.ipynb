```python
import numpy as np
import pandas as pd
d=pd.read_csv("DelayedFlights.csv")
```

```python
arr_delay=d['ArrDelay']
dep_delay=d['DepDelay']
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
plt.plot(dep_delay,arr_delay,'o')   
plt.ylabel("arr_delay")
plt.xlabel("dep_delay")
```

```python
mean_arr_delay=d['ArrDelay'].mean()
```

```python
plt.plot(dep_delay,arr_delay,'o')   # o is for plot style i.e circles #scatter plot
plt.ylabel("arr_delay")
plt.xlabel("dep_delay")
plt.axhline(mean_arr_delay,color='r',linestyle='-')
plt.show()
```
